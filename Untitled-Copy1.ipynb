```python
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

# Seed for reproducibility
np.random.seed(42)

# Generate patient IDs
num_patients = 100
patient_ids = [f"PID{1000+i}" for i in range(num_patients)]

# Generate records per patient (simulate 30 days)
records_per_patient = 30
total_records = num_patients * records_per_patient

data = []

for pid in patient_ids:
    base_date = datetime.now() - timedelta(days=30)
    
    age = np.random.randint(25, 80)
    gender = random.choice(['Male', 'Female'])
    has_disease = random.choice([0, 1])  # 0: No chronic disease, 1: Yes

    for i in range(records_per_patient):
        date = base_date + timedelta(days=i)
        
        # Simulated vitals with normal ranges
        pulse = np.random.normal(75, 10)
        spo2 = np.random.normal(97, 2)
        temp = np.random.normal(98.6, 1)
        bp_sys = np.random.normal(120, 10)
        bp_dia = np.random.normal(80, 5)

        # Inject anomalies randomly
        if np.random.rand() < 0.05:
            pulse += np.random.choice([-30, +40])
        if np.random.rand() < 0.03:
            spo2 -= np.random.choice([5, 10])
        if np.random.rand() < 0.03:
            temp += np.random.choice([2, 3])

        # Risk label logic: generate based on rules
        risk = int((pulse > 110 or spo2 < 90 or temp > 101 or has_disease == 1) and age > 50)

        data.append([
            pid, date, age, gender, has_disease,
            round(pulse, 1), round(spo2, 1), round(temp, 1),
            round(bp_sys, 1), round(bp_dia, 1),
            risk
        ])

# Create DataFrame
columns = ['patient_id', 'timestamp', 'age', 'gender', 'has_disease',
           'pulse', 'spo2', 'temperature', 'bp_sys', 'bp_dia', 'risk']

df = pd.DataFrame(data, columns=columns)

# Derive flags for rules engine
df['is_fever'] = df['temperature'] > 100.4
df['is_low_spo2'] = df['spo2'] < 94
df['is_high_bp'] = (df['bp_sys'] > 140) | (df['bp_dia'] > 90)

df['bp_diff'] = df['bp_sys'] - df['bp_dia']       # Pulse pressure
df['spo2_temp_ratio'] = df['spo2'] / df['temperature']  # Combined measure
df['is_elderly'] = (df['age'] >= 60).astype(int)   # Binary flag for age


# Save the simulated data
df.to_csv("synthetic_health_data.csv", index=False)
print("✅ Synthetic health data generated and saved!")

# View a few records
df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import joblib

# ---------- Load synthetic data ----------
df = pd.read_csv("synthetic_health_data.csv")

# ---------- Define features ----------
features = ['pulse', 'spo2', 'temperature', 'bp_sys', 'bp_dia']
X = df[features].copy()

# ---------- Handle missing values (safe and modern) ----------
X = X.ffill().bfill()

# ---------- Feature Scaling ----------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------- Train Isolation Forest ----------
iso_model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
iso_model.fit(X_scaled)

# ---------- Predict anomalies ----------
df['anomaly_label'] = iso_model.predict(X_scaled)  # -1 = anomaly, 1 = normal
df['anomaly'] = df['anomaly_label'].map({1: 'Normal', -1: 'Anomaly'})

# ---------- Save model and scaler ----------
joblib.dump(iso_model, 'anomaly_detector.joblib')
joblib.dump(scaler, 'anomaly_scaler.joblib')

# ---------- Preview ----------
print("✅ Anomaly detection model trained and saved.")
print(df[['pulse', 'spo2', 'temperature', 'bp_sys', 'bp_dia', 'anomaly']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = ['pulse', 'spo2', 'temperature', 'bp_sys', 'bp_dia', 'age', 'bp_diff', 'spo2_temp_ratio', 'is_elderly']
X = df[features]
y = df['risk']

# Handle missing values
X = X.ffill().bfill()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import joblib

# Simulate health risk data (replace with real patient time series if available)
np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=150, freq='D')
risk_scores = np.clip(np.random.normal(loc=0.6, scale=0.1, size=150), 0, 1)

df = pd.DataFrame({
    'date': dates,
    'risk': risk_scores
})

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Resample and take daily mean (real data may need this step)
daily_risk = df['risk'].resample('D').mean()

# Train Holt’s Exponential Smoothing model
model = ExponentialSmoothing(daily_risk, trend='add', seasonal=None)
fitted_model = model.fit()

# Save model and dataset
joblib.dump(fitted_model, 'risk_trend_model.joblib')
df.reset_index().to_csv('risk_data.csv', index=False)

print("Model trained and saved as 'risk_trend_model.joblib'")
```
